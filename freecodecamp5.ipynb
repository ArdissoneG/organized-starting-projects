```python
import pandas as pd
df = pd.read_csv(r'C:\Users\Personal\Desktop\programación\csv que voy usando/epa-sea-level.csv')
df
```

```python
import matplotlib.pyplot as plt
df.plot.scatter(x='Year', y='CSIRO Adjusted Sea Level', s=1, c= 'Black')
```

```python
from scipy.stats import linregress
import numpy as np

x = df['Year']
y = df['CSIRO Adjusted Sea Level']

res = linregress(x,y)

x_pred = np.arange(x.min(), 2051)
y_pred = res.intercept + res.slope * x_pred

plt.figure(figsize=(10, 6))
plt.scatter(x, y, s=1, c='black')
plt.plot(x_pred, y_pred, 'r', label='Linea de regresión')
plt.xlim(x.min(), 2051)
plt.xlabel("Year")
plt.ylabel("Sea Level")
plt.title("Rise in Sea Level")
plt.legend()
plt.show()
```
