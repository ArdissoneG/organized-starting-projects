```python
import pandas as pd

df = pd.read_csv(r'C:\Users\Personal\Desktop\programación/csv que voy usando/medical_examination (1).csv')
df['age']=df['age']//365 #edad estaba en días
df
```

```python
df['overweight'] = ((df['weight']/((df['height']/100)**2))>25).astype(int)
df
```

```python
df['cholesterol'] = df['cholesterol'].map({1: 0, 2: 1, 3: 1})
df['gluc'] = df['gluc'].map({1: 0, 2: 1, 3: 1})
df
```

```python
df_cat = pd.melt(
        df,id_vars=['cardio'],
        value_vars=['cholesterol', 'gluc', 'smoke', 'alco', 'active', 'overweight'])
df_cat
```

```python
df_cat2 = df_cat.groupby(['cardio', 'variable', 'value']).size().reset_index(name='total')
df_cat2
```

```python
import seaborn as sns
fig = sns.catplot(
    x='variable',
    y='total',
    hue='value',
    col='cardio',
    data=df_cat2,
    kind='bar'
).fig
```

```python
import numpy as np
import matplotlib.pyplot as plt
df_heat = df[
    (df['ap_lo'] <= df['ap_hi']) &
    (df['height'] >= df['height'].quantile(0.025)) &
    (df['height'] <= df['height'].quantile(0.975)) &
    (df['weight'] >= df['weight'].quantile(0.025)) &
    (df['weight'] <= df['weight'].quantile(0.975))
]

corr = df_heat.corr(numeric_only=True)
mask = np.triu(np.ones_like(corr, dtype=bool))
fig, ax = plt.subplots(figsize=(12, 10))
sns.heatmap(
    corr,
    mask=mask,
    annot=True,
    fmt=".1f",
    center=0,
    vmax=0.3,
    square=True,
    linewidths=0.5,
    cbar_kws={"shrink": 0.5}
)
```
